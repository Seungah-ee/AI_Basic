```python
import numpy as np   
import matplotlib.pyplot as plt

x = 2
y = 2
b = 0

w = np.linspace(-1, 3, 50)
# print(w)
j = np.zeros(len(w))
j[0] += 3
# print(j)

# for i in range(50):
#     print(i)

for i in range(len(w)):
    y_hat = w[i]*x+b
    j[i] = 0.5*(y_hat - y)**2
# print(j)
plt.plot(w, j)
plt.plot(w, j, 'ro')
plt.grid(True)
plt.show()
```
